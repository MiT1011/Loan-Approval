```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import svm
```

```python
#overview of the data
df = pd.read_csv("loan1.csv")
df.head()
df
```

```python
#Find the missing values of the data
df.isnull().sum()
```

```python
df['loanAmount_log'] = np.log(df['LoanAmount']) #it creates new column named with loanAmount_log with logarithm of LoagAmount with base e
df['loanAmount_log'].hist(bins = 20)
```

```python
df.isnull().sum()
```

```python
df['TotalIncome'] = df['ApplicantIncome'] + df ['CoapplicantIncome'] #Creating new column named with TotalIncome
df['TotalIncome_log'] = np.log(df['TotalIncome'])
df['TotalIncome_log'].hist(bins=20)
```

```python
df['Gender'].fillna(df['Gender'].mode()[0], inplace = True) #inplace means it will not creat new data frame it will change in existing
df['Married'].fillna(df['Married'].mode()[0], inplace = True) #mode()[0] this will choose first mode value
df['Self_Employed'].fillna(df['Self_Employed'].mode()[0], inplace = True)
df['Dependents'].fillna(df['Dependents'].mode()[0], inplace = True)

df.LoanAmount = df.LoanAmount.fillna(df.LoanAmount.mean())
df.loanAmount_log = df.loanAmount_log.fillna(df.loanAmount_log.mean())

df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0], inplace = True)
df['Credit_History'].fillna(df['Credit_History'].mode()[0], inplace = True)

df.isnull().sum()
```

```python
#we will select specific rows and colums for training and testing
x = df.iloc[:,np.r_[1:5,9:11,13:15]].values
y = df.iloc[:,12].values

x
```

```python
#Training of the dataset
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder #encode categorical labels into numerical values.

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.2, random_state = 0) #X_train is for training, X_test for testing, Y_train and Y_test for for target sets
Labelencoder_x = LabelEncoder()
```

```python
for i in range(0,5):
    X_train[:,i] = Labelencoder_x.fit_transform(X_train[:,i])
    X_train[:,7] = Labelencoder_x.fit_transform(X_train[:,7])

X_train
```

```python
Labelencoder_y = LabelEncoder()
Y_train = Labelencoder_y.fit_transform(Y_train)

Y_train
```

```python
#Testing of the data
for i in range(0,5):
    X_test[:,i] = Labelencoder_x.fit_transform(X_test[:,i])
    X_test[:,7] = Labelencoder_x.fit_transform(X_test[:,7])

X_test
```

```python
Labelencoder_y = LabelEncoder()

Y_test = Labelencoder_y.fit_transform(Y_test)

Y_test
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.fit_transform(X_test)
```

```python
#Random Forest
#1 - Loan Aprrove
#2 - Loan not Approve

from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier()
rf_clf.fit(X_train, Y_train)
```

```python
from sklearn import metrics

y_pred = rf_clf.predict(X_test)

print("Accuracy of the random forest classifier is", metrics.accuracy_score(y_pred, Y_test))

y_pred
```

```python
# Naive Bayes Theorem
from sklearn.naive_bayes import GaussianNB
nb_clf = GaussianNB()
nb_clf.fit(X_train, Y_train)
```

```python
y_pred = nb_clf.predict(X_test)
print("Accuracy of naive bayes classifier is", metrics.accuracy_score(y_pred, Y_test))
```

```python
y_pred
```

```python
#Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier

dt_clf = DecisionTreeClassifier()
dt_clf.fit(X_train, Y_train)
```

```python
y_pred = dt_clf.predict(X_test)
print("Accuracy of Decision Tree classifier is", metrics.accuracy_score(y_pred, Y_test))
```

```python
y_pred
```

```python
#knn
from sklearn.neighbors import KNeighborsClassifier
kn_clf = KNeighborsClassifier()
kn_clf.fit(X_train, Y_train)
```

```python
y_pred = kn_clf.predict(X_test)
print("Accuracy of Knn is", metrics.accuracy_score(y_pred, Y_test))
```
